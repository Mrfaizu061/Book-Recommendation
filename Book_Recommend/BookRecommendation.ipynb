```python
#now we have to filter those users who rated atleast 200 books

# Now x[x]

# Here x on the right-hand side is used as a mask.
# It keeps only the rows where the value is True.
# index is used for the getting users ids
x = rating_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_linke_users=x[x].index
padhe_linke_users
```

```python
# 811 is user who has rated the altease 200 books
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(padhe_linke_users)]
filtered_rating
```

```python
#This will return a DataFrame with only the ratings given by your active users (≥200 ratings).
```

```python
# Step 1: Count how many ratings each book got
# Step 2: Keep only True values
# y[y]

# Step 3: Get just the book titles
# famous_books = y[y].index

y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
# ✅ famous_books is now an Index object containing only the titles of books with ≥ 50 ratings:
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
# there is 810 users(padhe likhe user) and 706 rows
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Cosine similarity measures how similar two vectors are, by looking at the angle between them instead of their raw values.
# If two vectors point in the same direction → cosine similarity = 1 (perfectly similar).
# If they are 90° apart → cosine similarity = 0 (no similarity).
# If they point in opposite directions → cosine similarity = -1 (completely opposite).
```

```python
similarity_score=cosine_similarity(pt)
similarity_score.shape
```

```python
# RECOMMENDATION 
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)

    return data;
```

```python
recommend('The Notebook')
```
